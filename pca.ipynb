```python
from __future__ import print_function, division

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import SparsePCA
```

```python
data = pd.read_csv('data_pca_joined.csv')
by = pd.read_csv('new_gapdata.csv')
df_by = by[['YearBuilt', 'Full_Address']]
```

```python
data = pd.merge(data, df_by, how='left', on='Full_Address')
```

```python
data = data.dropna()
```

```python
data['BldgAge'] = 2016-data['YearBuilt']
```

```python
data_model = data[['normalized_gap', 'commute_up60min_ratio', 'Household_65over_ratio', 'FuelOil_ratio', 'Spanish_lan_ratio', 'Asian_lan_ratio', 'BldgAge']]
```

```python
#NormalPCA
pca = PCA(n_components=2)
X_proj = pca.fit_transform(values[:,0:25])
columns = ['1_value','2_value', 'feature']
df_pca = pd.DataFrame(columns=columns)
df_pca.feature = data.columns[0:25]
df_pca['1_value'] = pca.components_[0,:]
df_pca['2_value'] = pca.components_[1,:]
```
